```python
# These are the different imports used in the code.

import pandas as pd

import os
```

```python
#Estract the playes

def extract_player_info(df):
    # Empty json object to store the player information
    players = []
    teams = []
    teams.append(df.iloc[2,3])
    teams.append(df.iloc[2,4])
    for team in teams:
        keys = team.keys()
        for key in keys:
            player_info = team[key]['info']
            players.append({'playerid':player_info['idplayer'],'lastname': player_info['lastname']})
    return players
```

```python
#Extract the team information


def extract_team_info(df):
    # Empty json object to store the player information
    teamsinfo = []
    teamnames = []
    teams = []
    teams.append(df.iloc[2,3])
    teams.append(df.iloc[2,4])
    teamnames.append(df.iloc[1,3])
    teamnames.append(df.iloc[1,4])
    for team, name in zip(teams, teamnames):
        keys = team.keys()
        for key in keys:
            player_info = team[key]['info']
            teamsinfo.append({'idteam':player_info['idteam'], 'name':name})
    return teamsinfo
```

```python
# Get the list of all files and directories
path = "data"
dir_list = os.listdir(path)
dir_list
```

```python
players_data = []
teams_data = []

for file in dir_list:
    df = pd.read_json('data/' + file)
    extracted_values = extract_player_info(df)
    extracted_teams = extract_team_info(df)
    teams_data.extend(extracted_teams)
    players_data.extend(extracted_values)

len(players_data)
```

```python
unique_players = []

for player in players_data:
    if player not in unique_players:
        unique_players.append(player)
len(unique_players)
```

```python
# Remove duplicates from the list of teams
unique_team = []

for team in teams_data:
    if team not in unique_team:
        unique_team.append(team)

unique_team

```

```python
# Create a dataframe from the list of unique players
players_df = pd.DataFrame(unique_players)

players_df
```

```python
# Create a dataframe from the list of unique teams
team_df = pd.DataFrame(unique_team)

team_df
```
