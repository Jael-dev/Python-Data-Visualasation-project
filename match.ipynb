```python
teamsinfo = []
teamnames = []
def extract_team_info(df):
    # Empty json object to store the player information
    teams = []
    teams.append(df.iloc[2,3])
    teams.append(df.iloc[2,4])
    teamnames.append(df.iloc[1,3])
    teamnames.append(df.iloc[1,4])
    for team, name in zip(teams, teamnames):
        keys = team.keys()
        for key in keys:
            player_info = team[key]['info']
            teamsinfo.append({'idteam':player_info['idteam'], 'name':name})
    return teamsinfo
```

```python
df = pd.read_json('data/match_2137741.json')

extract_team_info(df)
```

```python
# Remove duplicates from the list
unique_team = []

for team in teamsinfo:
    if team not in unique_team:
        unique_team.append(team)

unique_team

```
